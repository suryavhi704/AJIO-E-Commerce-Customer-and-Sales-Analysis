# 1. What are the top-selling product categories and brands?

```python
products.head(2)
```

```python
# merging of dataframes:
ordered_product=pd.merge(orders,products,how='inner',on='P_ID')
```

```python
ordered_product['Category'].value_counts()
```

```python
# What is the top selling product name:-
top_sell_products=ordered_product.groupby('Category')['Qty'].sum().reset_index()
```

```python
sns.barplot(data=top_sell_products,x='Category',y='Qty', palette='rocket')
plt.xticks(rotation=45)
plt.title('Top Selling Product Categories by Quantity')
plt.xlabel('Product Category')
plt.ylabel('Quantity Sold')
plt.tight_layout()
plt.show()
```

```python
# Top selling company:
top_selling_company=ordered_product['Company_Name'].value_counts().reset_index()
```

```python
top_selling_company
```

```python
sns.barplot(data=top_selling_company,x='Company_Name',y='count',palette='Greys')
plt.title('Top Selling Brand')
plt.xticks(rotation=90)
plt.show()
```

Highest selling company is puma and highest selling category is blazer.

# 4.What is the return rate of products, and which categories are returned most often?

```python
returns.head()
```

```python
orders.head()
```

```python
new_returns=returns[['RT_ID','Or_ID']]
```

```python
new_returns.head()
```

```python
return_details=orders.merge(new_returns,how='left',on='Or_ID')
return_details.head()
```

```python
not_returned=return_details.isnull().sum().sum()
```

```python
returned_orders=return_details[~return_details['RT_ID'].isnull()]
```

```python
returned_orders # orders which have been returned.
```

```python
total=return_details.shape[0]
```

```python
returned=total-not_returned
returned
```

```python
print('Rate of return:-',(returned/total)*100)
```

Out of 10.18k oreders 2k (19.65%) orders are returned.

```python
products.head()
```

```python
updated_products=products[['P_ID','Category']]
updated_products.head()
```

```python
returned_category=returned_orders.merge(updated_products,on='P_ID',how='inner')
returned_category
```

```python
result=returned_category['Category'].value_counts(normalize=True).reset_index()
result
```

```python
sns.barplot(data=result,x='Category',y='proportion',palette='Blues')
plt.xlabel('Categories')
plt.title('Most Returned Categories')
plt.ylabel('Percentage of Return')
plt.show()
```

Most returned categories are shirt,blazer,sweater.
